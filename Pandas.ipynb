# 함수 적용

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
```

## 데이터에 함수 적용 (apply)

```python
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```

```python
df['SW특기'].str.capitalize()
```
